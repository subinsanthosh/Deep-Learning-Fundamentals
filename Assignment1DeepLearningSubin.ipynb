```python
import torch
import torch.nn as nn
import torch.optim as optim
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
y = np.where(data['Outcome'] == 0, -1, 1) # convert labels from (0,1) to (-1, +1) for the perceptron
X = data.drop(columns=['Outcome']).values
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y) # split the data into training 70% and testing 30% set

scaler = StandardScaler() # standardizing the features
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train = torch.tensor(X_train, dtype=torch.float32)
y_train = torch.tensor(y_train, dtype=torch.float32)
X_test = torch.tensor(X_test, dtype=torch.float32)
y_test = torch.tensor(y_test, dtype=torch.float32) # convert data into PyTorch tensor
```

# Perceptron model

```python
n_features = X_train.shape[1] # initialize weights and bias
weights = torch.zeros(n_features, dtype=torch.float32, requires_grad=False)
bias = torch.rand(1, dtype=torch.float32, requires_grad=False)

learning_rate = 0.001  # learning rate for better convergence
n_epochs = 2000  # epochs for better training

train_misclassifications = []
val_misclassifications = []

patience = 100  # number of epochs to wait for improvement
min_val_misclassified = np.inf
best_epoch = 0

# training loop with Perceptron update rule and early stopping
for epoch in range(n_epochs):
    misclassified_count = 0

    perm = torch.randperm(X_train.size(0))
    X_train_shuffled = X_train[perm]
    y_train_shuffled = y_train[perm]

    for i in range(X_train_shuffled.size(0)):
        x_i = X_train_shuffled[i]
        y_i = y_train_shuffled[i]

        linear_output = torch.dot(weights, x_i) + bias 
        condition = y_i * linear_output # calculate the dot product


        if condition <= 0:
            weights += learning_rate * y_i * x_i  # update weights if the sample is misclassified
            bias += learning_rate * y_i           # update bias
            misclassified_count += 1
    train_misclassifications.append(misclassified_count)

    # validation phase to check misclassifications
    val_misclassified_count = 0
    for i in range(X_test.size(0)):
        x_i = X_test[i]
        y_i = y_test[i]

        linear_output = torch.dot(weights, x_i) + bias
        prediction = 1 if linear_output >= 0 else -1

        if prediction != y_i:
            val_misclassified_count += 1

    val_misclassifications.append(val_misclassified_count)

    if val_misclassified_count < min_val_misclassified:
        min_val_misclassified = val_misclassified_count
        best_epoch = epoch
        best_weights = weights.clone()
        best_bias = bias.clone()

    if (epoch - best_epoch) >= patience:
        print(f"Early stopping at epoch {epoch + 1}. Best epoch was {best_epoch + 1} with validation misclassified: {min_val_misclassified}")
        weights = best_weights
        bias = best_bias
        break
```

```python
plt.figure(figsize=(12, 8))
plt.plot(range(1, len(train_misclassifications) + 1), train_misclassifications, label='Training Misclassifications')
plt.plot(range(1, len(val_misclassifications) + 1), val_misclassifications, label='Validation Misclassifications', linestyle='--')
plt.xlabel('Epochs')
plt.ylabel('Number of Misclassifications')
plt.title('Learning Curves: Perceptron Model')
plt.legend()
plt.grid(True)
plt.show()
```

```python
y_pred = []
for i in range(X_test.size(0)):
    x_i = X_test[i]
    linear_output = torch.dot(weights, x_i) + bias
    prediction = 1 if linear_output >= 0 else -1
    y_pred.append(prediction)

y_pred_np = np.array(y_pred)
y_test_np = y_test.numpy()
```

```python
accuracy = accuracy_score(y_test_np, y_pred_np)
print(f'Test Accuracy: {accuracy:.4f}')
```

```python
classification_rep = classification_report(y_test_np, y_pred_np, target_names=['Non-Diabetic', 'Diabetic'])
print('Classification Report:')
print(classification_rep)
```

```python
conf_matrix = confusion_matrix(y_test_np, y_pred_np)

plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues', xticklabels=['Non-Diabetic', 'Diabetic'], yticklabels=['Non-Diabetic', 'Diabetic'])
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix - Perceptron Model')
plt.show()
```

```python
from sklearn.metrics import roc_curve, auc, precision_recall_curve
```

```python
fpr, tpr, _ = roc_curve(y_test_np, y_pred_np)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()
```

```python
precision, recall, _ = precision_recall_curve(y_test_np, y_pred_np)

plt.figure(figsize=(8, 6))
plt.plot(recall, precision, color='b', lw=2, label='Precision-Recall curve')
plt.xlabel('Recall')
plt.ylabel('Precision')
plt.title('Precision-Recall Curve')
plt.legend(loc="lower left")
plt.grid(True)
plt.show()
```
